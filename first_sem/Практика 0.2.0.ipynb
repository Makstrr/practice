Задание 7: Необходимо создать простой калькулятор, который позволяет пользователю выбрать одну из четырёх операций (+, -, *, /), ввести два числа и получить результат выполнения операции.

`Введите операцию:`

`Введите первое число:`

`Введите второе число:`

`Результат выполнения: `


```python
operator = input('Введите операцию: ')
first_num = int(input('Введите первое число: '))
second_num = int(input('Введите второе число: '))
match operator:
  case '+':
    print('Результат выполнения:', first_num+second_num)
  case '-':
    print('Результат выполнения:', first_num-second_num)
  case '*':
    print('Результат выполнения:', first_num*second_num)
  case '/':
    print('Результат выполнения:', first_num/second_num)
```

# Практика 0.2.1

**Шахматы**

Даны стартовые и конечные координаты, а также фигура

Необходимо определить, может ли заданная фигура так ходить?

```python
start_coords = list(map(int, input('Введите начальные координаты: ').split(', ')))
end_coords = list(map(int, input('Введите конечные координаты: ').split(', ')))
figure = input('Введите название фигуры: ')
figure = figure.lower()
answer = ['Не может', 'Может']
match figure:
  case 'пешка':
    flag = start_coords[0] == end_coords[0] and start_coords[1] + 1 == end_coords[1]
    print(answer[flag])
  case 'ладья':
    flag = (start_coords[0] == end_coords[0] and start_coords[1] != end_coords[1]) or (start_coords[1] == end_coords[1] and start_coords[0] != end_coords[0])
    print(answer[flag])
  case 'конь':
    flag = (abs(start_coords[0] - end_coords[0]) == 1 and abs(start_coords[1] - end_coords[1]) == 2) or (abs(start_coords[1] - end_coords[1]) == 1 and abs(start_coords[0] - end_coords[0]) == 2)
    print(answer[flag])
  case 'слон':
    flag = abs(start_coords[0] - end_coords[0]) == abs(start_coords[1] - end_coords[1])
    print(answer[flag])
  case 'король':
    flag = abs(start_coords[0] - end_coords[0]) <= 1 and abs(start_coords[1] - end_coords[1]) <= 1
    print(answer[flag])
  case 'ферзь':
    flag = (abs(start_coords[0] - end_coords[0]) == abs(start_coords[1] - end_coords[1])) or ((start_coords[0] == end_coords[0] and start_coords[1] != end_coords[1]) or (start_coords[1] == end_coords[1] and start_coords[0] != end_coords[0]))
    print(answer[flag])
```

# Практика 0.2.2

**Цифра на определенном месте:**

Последовательно записан натуральный ряд чисел.

Какая цифра стоит в N позиции

```python
#index = int(input('Введите число N: '))
#print('В натуральном ряде чисел на %d-ой позиции стоит число %d' % (index, index))
def find_digit(n):
  number_length = 1
  count_of_numbers = 9
  start_number = 1

  while n > number_length * count_of_numbers:
    n -= number_length * count_of_numbers
    number_length += 1
    count_of_numbers *= 10
    start_number *= 10

  number_index = (n-1) // number_length
  digit_index = (n-1) % number_length
  return str(start_number+number_index)[digit_index]


N = int(input('Введите число N: '))
print('В натуральном ряде чисел на %d-ой позиции стоит цифра %s' % (N, find_digit(N)))
```
