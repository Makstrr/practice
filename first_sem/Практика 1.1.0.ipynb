# Задание 1

**Описание:** Создайте иерархию классов для разных типов сотрудников в компании. Реализуйте родительский класс Employee и дочерние классы Manager и Developer. Каждый класс должен иметь метод для расчета зарплаты на основе различных критериев класса.


Отрабатываемый принцип: Наследование



```python
class Employee:

  def __init__(self, name, experience, working_hours):
    self.name = name
    self.experience = experience
    self.working_hours = working_hours

  def get_info(self):
    info = ('Имя: ' + self.name +
    '\nСтаж: ' + str(self.experience) +
    '\nКол-во рабочих часов: ' + str(self.working_hours))
    return info


class Manager(Employee):

  def salary_estimation(self):
    salary = (100 * self.working_hours * (1 + self.experience*0.05)) * 4
    return salary


class Developer(Employee):

  def salary_estimation(self):
    salary = (200 * self.working_hours * (1 + self.experience*0.08)) * 4
    return salary


manager = Manager('Иван', 5, 30)
print(manager.get_info())
print('Зарплата:', manager.salary_estimation())
print()

developer = Developer('Роберт', 10, 40)
print(developer.get_info())
print('Зарплата:', developer.salary_estimation())
```

# Задание 3

Онлайн-магазин:
- Создайте модель для онлайн-магазина с классами Product, Order, Customer, и ShoppingCart.
- Product включает информацию о цене, наличии на складе и категории товара.
Order обрабатывает процесс покупки, включая расчет цены с учетом скидок и налогов.
- Customer управляет информацией о пользователе и его истории заказов.
- ShoppingCart позволяет добавлять, удалять и обновлять количество товаров перед оформлением заказа.


```python
class Product:

    def __init__(self, name, price, stock, category):
        self.name = name
        self.price = price
        self.stock = stock
        self.category = category

    def is_available(self, quantity):
        return self.stock >= quantity


class Order:

    def __init__(self):
        self.items = []
        self.total = 0.0

    def add_item(self, product, quantity):
        if product.is_available(quantity):
            product.stock -= quantity
            self.items.append((product, quantity))
            self.calculate_total()
        else:
            print(f"Недостаточно товара {product.name} на складе.")

    def calculate_total(self, discount=0, tax=0):
        subtotal = sum(item[0].price * item[1] for item in self.items)
        self.total = subtotal - (subtotal * discount) + (subtotal * tax)


class Customer:

    def __init__(self, name, email):
        self.name = name
        self.email = email
        self.order_history = []

    def add_order(self, order):
        self.order_history.append(order)


class ShoppingCart:

    def __init__(self):
        self.cart_items = []

    def add_product(self, product, quantity):
        self.cart_items.append((product, quantity))

    def remove_product(self, product):
        self.cart_items = [(p, q) for p, q in self.cart_items if p != product]

    def update_quantity(self, product, quantity):
        for idx, (p, q) in enumerate(self.cart_items):
            if p == product:
                self.cart_items[idx] = (p, quantity)
                break

    def checkout(self, customer):
        order = Order()
        for product, quantity in self.cart_items:
            order.add_item(product, quantity)
        customer.add_order(order)
        self.cart_items.clear()
        return order


product1 = Product("Телефон", 500, 10, "Электроника")
product2 = Product("Наушники", 100, 5, "Аудио")
customer = Customer("Иван", "ivan@example.com")
shopping_cart = ShoppingCart()

shopping_cart.add_product(product1, 1)
shopping_cart.add_product(product2, 2)

order = shopping_cart.checkout(customer)
print(f"Заказ на сумму: {order.total}")
```

# Задание 4

Симулятор космического корабля:
- Создайте симулятор управления космическим кораблем с классами SpaceShip, CrewMember, и Mission.
- SpaceShip имеет атрибуты для управления топливом, состоянием корпуса, и текущей скоростью.
- CrewMember контролирует здоровье, навыки, и роли в команде (например, пилот, инженер).
- Mission определяет цели, ресурсы, и возможные события (например, аварии, встречи с астероидами).

```python
from random import randint


class SpaceShip:

  def __init__(self, fuel, durability, current_velocity):
    self.fuel = fuel
    self.durability = durability
    self.current_velocity = current_velocity

  def accelerate(self, crewMember):
    if crewMember.role == 'pilot':
      self.current_velocity += 100
      self.fuel -= 1
      print('Скорость повышена на 100')
    else:
      print('Выбранный член экипажа не обладает необходимой квалификацией')

  def repair(self, crewMember):
    if crewMember.role == 'engineer':
      self.durability += 100 * (1 + crewMember.skill/100)
      crewMember.skill += 1
      print('Состояние корпуса повышено на 100')
    else:
      print('Выбранный член экипажа не обладает необходимой квалификацией')

  def refuel(self, crewMember):
    if crewMember.role == 'engineer':
      self.fuel = 100
      print('Топливо пополнено')
    else:
      print('Выбранный член экипажа не обладает необходимой квалификацией')


class CrewMember:

  def __init__(self, name, health, skill, role):
    self.name = name
    self.health = health
    self.skill = skill
    self.role = role

  def heal(self):
    self.health += 10

  def aid(self, crewMember):
    if self.name != crewMember.name:
      crewMember.health += 50


class Medic(CrewMember):

  def __init__(self, name, health, skill):
    self.name = name
    self.health = health
    self.skill = skill
    self.role = 'medic'

  def aid(self, crewMember):
    if self.name != crewMember.name:
      crewMember.health += 200 * (1 + self.skill/100)
      self.skill += 1


class Scientist(CrewMember):

  def __init__(self, name, health, skill):
    self.name = name
    self.health = health
    self.skill = skill
    self.role = 'scientist'

  def research(self, mission):
    mission.progress += 5 * (1 + self.skill/100)
    self.skill += 1


class Mission:

  def __init__(self, category, name, spaceShip):
    self.progress = 0
    self.category = category
    self.name = name
    self.spaceShip = spaceShip
    self.isEnded = False

  def activity(self):
    if not self.isEnded:
      match self.category:
        case 'malfunction':
          self.spaceShip.current_velocity = 0
          self.progress += 25
          if self.progress >= 100:
            self.isEnded == True
        case 'asteroids_collision':
          self.spaceShip.durability -= 50
          self.progress += 50
          if self.progress >= 100:
            self.isEnded == True
        case 'research':
          if self.progress >= 100:
            self.isEnded == True


def main():
  ishimura = SpaceShip(100, 2000, 40)
  missions = []
  crewMembers = []
  print('Добавьте членов экипажа')
  while True:
    role = input('Роль члена экипажа: ').lower()
    health = int(input('Здоровье члена экипажа: '))
    skill = int(input('Уровень навыков члена экипажа: '))
    name = input('Имя члена экипажа: ')
    if role == 'medic':
      crewMembers.append(Medic(name, health, skill))
    if role == 'scientist':
      crewMembers.append(Scientist(name, health, skill))
    else:
      crewMembers.append(CrewMember(name, health, skill, role))
    temp = input('Продолжить? ')
    if temp.lower() != 'да':
      break
  print()
  print('Доступные команды:',
        '\n1 - открыть бортовой журнал',
        '\n2 - открыть список экипажа',
        '\n3 - действия экипажа',
        '\n0 - выход\n')
  while True:
    choice = input()
    print()
    match choice:
      case '1':
        print('\nТопливо:', ishimura.fuel,
              '\nСостояние корпуса:', ishimura.durability,
              '\nТекущая скорость:', ishimura.current_velocity,
              '\n')
        for mission in missions:
          print('\nТип:', mission.category,
                '\nПрогресс:', mission.progress,
                '\n')
      case '2':
        for member in crewMembers:
          print('\nИмя:', member.name,
                '\nРоль:', member.role,
                '\nЗдоровье:', member.health,
                '\nУровень навыков:', member.skill,
                '\n')
      case '3':
        print('\nВы можете',
              '\n1 - ускорить корабль',
              '\n2 - пополнить топливо',
              '\n3 - восстановить корпус',
              '\n4 - самопомощь члена экипажа',
              '\n5 - оказать помощь одному члену экипажа другим\n',
              '\n6 - продолжить исследование')
        action_choice = input()
        match action_choice:
          case '1':
            member_choice = int(input('Введите номер члена экипажа: ')) - 1
            ishimura.accelerate(crewMembers[member_choice])
          case '2':
            member_choice = int(input('Введите номер члена экипажа: ')) - 1
            ishimura.refuel(crewMembers[member_choice])
          case '3':
            member_choice = int(input('Введите номер члена экипажа: ')) - 1
            ishimura.accelerate(crewMembers[member_choice])
          case '4':
            member_choice = int(input('Введите номер члена экипажа: ')) - 1
            crewMembers[member_choice].heal()
          case '5':
            member_choice1=(int(input('Введите номер первого члена экипажа: '))
            - 1)
            member_choice2=(int(input('Введите номер второго члена экипажа: '))
            - 1)
            crewMembers[member_choice1].aid(crewMembers[member_choice2])
          case '6':
            member_choice = int(input('Введите номер члена экипажа: ')) - 1
            mission_choice = int(input('Введите номер миссии: ')) - 1
            crewMembers[member_choice].research(missions[mission_choice])
          case _:
            print()
      case '0':
        break
      case _:
        print('Неверный ввод')
    for mission in missions:
      mission.activity()
    if randint(0, 100) > 70:
      newMission = randint(0, 2)
      match newMission:
        case 0:
          mission.append(Mission('malfunction', 'Неисправность систем корабля',
                                ishimura))
        case 1:
          missions.append(Mission('asteroids_collision',
                                  'Столкновение с астероидами', ishimura))
        case 2:
          missions.append(Mission('research', 'Исследование', ishimura))


main()
```

**Описание:** создайте консольную версию игры крестики-нолики, используя классы

```python
class TicTacToe:
    def __init__(self):
        self.board = [' ' for _ in range(9)]
        self.current_player = 'X'

    def print_board(self):
        print(f"{self.board[0]} | {self.board[1]} | {self.board[2]}")
        print("--|---|--")
        print(f"{self.board[3]} | {self.board[4]} | {self.board[5]}")
        print("--|---|--")
        print(f"{self.board[6]} | {self.board[7]} | {self.board[8]}")

    def make_move(self, position):
        if self.board[position] == ' ':
            self.board[position] = self.current_player
            return True
        else:
            print("Эта клетка уже занята.")
            return False

    def switch_player(self):
        self.current_player = 'O' if self.current_player == 'X' else 'X'

    def check_winner(self):
        winning_combinations = [
            (0, 1, 2), (3, 4, 5), (6, 7, 8),
            (0, 3, 6), (1, 4, 7), (2, 5, 8),
            (0, 4, 8), (2, 4, 6)
        ]

        for combo in winning_combinations:
            if (self.board[combo[0]] == self.board[combo[1]] ==
             self.board[combo[2]] != ' '):
                return self.board[combo[0]]
        return None

    def is_draw(self):
        return ' ' not in self.board


def main():
    game = TicTacToe()

    while True:
        game.print_board()
        try:
            move = int(input(f"Игрок {game.current_player}, " +
                             f"выберите позицию (0-8): "))
            if move < 0 or move > 8:
                print("Пожалуйста, введите номер от 0 до 8.")
                continue

            if game.make_move(move):
                winner = game.check_winner()
                if winner:
                    game.print_board()
                    print(f"Поздравляем, игрок {winner} выиграл!")
                    break
                elif game.is_draw():
                    game.print_board()
                    print("Игра завершилась вничью!")
                    break
                game.switch_player()
        except ValueError:
            print("Пожалуйста, введите корректное число.")


main()
```
