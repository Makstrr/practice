# **Задача 1.**




Типы данных Напишите программу на Python, которая принимает список чисел в качестве входных данных от пользователя, а затем выводит сумму, среднее значение и максимальное значение из списка.

***Нельзя использовать: len(), min(), max(), average(), sum()***

Ввод:
```
1 2 3 4 5 6 7
```
Вывод:

```
Сумма чисел: 28
Среднее арифм.: 4.0
Максимальное значение: 7
```



````python
nums = list(map(int, input().split(' ')))
length = 0
max_num = float('-inf')
nums_sum = 0
for i in nums:
  length += 1
  nums_sum += i
  if i > max_num:
    max_num = i
avg = nums_sum/length
print('Сумма чисел: %d\nСреднее арифм.: %f\nМаксимальное значение: %d' % (nums_sum, avg, max_num))
````

# **Задача 3**

Простой чат-бот
Создайте простого чат-бота, который отвечает на основные запросы пользователей. Программа должна использовать условные выражения для определения намерений пользователя и последующего соответствующего ответа.

Чат бот может:

- помогать складывать числа
- подсказывать фильм, музыку и тд (из заранее созданного списка)
- создавать список дел (а также выводить его, обновлять, очищать)

Интерфейс чат-бота на ваше усмотрение, это могут быть как заранее определенные команды, так и к примеру вызов команды по номеру

*Реализовать без использования функций*

Пример:

Ввод:

```
Олег, какую мызыку посоветуешь?
```

Вывод:

```
Я бы посоветовал вам: Led Zeppelin Stairway to Heaven
```

````python
from random import randint


print('Привет, я - чат-бот. Доступные команды:\n1 - складывание двух чисел\n2.1 - подскажу фильм\n2.2 - подскажу музыку\n3 - перейти в интерфейс списка дел\n0 - выход\n')
todo_list = []
running = 1
music = ['Metallica', 'Avenged Sevenfold', 'Megadeth', 'Five Fingers Death Punch', 'Exumer', 'Sum 41', 'Anthrax', 'Motley Crue', 'Papa Roach', 'Iron Maiden', 'Thousand Foor Krutch', 'Three Days Grace', 'Billy Talent', 'Static-X', 'Гражданская Оборона', 'Кино', 'Агата Кристи', 'Смысловые галлюцинации']
movies = ["Адвокат дьявола", "Титаник", "Матрица", "Терминатор", "Драйв", "Константин: повелитель тьмы"]
while running:
  choice = input()
  match choice:
    case '0':
      running = 0
    case '1':
      print('Введи два числа через пробел')
      nums = list(map(int, input().split(' ')))
      print('Сумма введенных чисел:', sum(nums), '\n')
    case '2.1':
      print('Я бы посоветовал', movies[randint(0, len(movies)-1)], '\n')
    case '2.2':
      print('Я бы посоветовал', music[randint(0, len(music)-1)], '\n')
    case '3':
      todo_running = 1
      while todo_running:
        print('Что ты хочешь сделать?\n1 - вывести список дел\n2 - добавить дело в список\n3 - удалить дело под номером n из списка\n4 - полностью очистить список\n0 - выйти в основной интерфейс\n')
        choice_todo = input()
        match choice_todo:
          case '1':
            if len(todo_list):
              print(*todo_list, sep='\n', end='\n\n')
            else:
              print('Список дел пуст\n')
          case '0':
            todo_running = 0
          case '4':
            todo_list = []
            print('Список дел очищен\n')
          case '3':
            n = int(input('Введи номер дела, которое хочешь удалить: '))
            del todo_list[n-1]
            print('Дело %d удалено\n' % n)
          case '2':
            temp = input('Введи название дела: ')
            todo_list.append(str(len(todo_list)+1) + '. ' + temp)
            print('Дело добавлено!\n')
          case _:
            print('Некорректная команда, повтори ввод\n')
    case _:
      print('Некорректная команда, повтори ввод\n')
````

# **Задача 4.**


Камень, ножницы, бумага

Создайте игру в Камень, ножницы, бумагу, в которой пользователь может играть против компьютера. Программа должна запросить выбор пользователя, а затем сгенерировать выбор компьютера. Затем программа должна определить победителя на основе правил игры.

**Этап 1:**

Определение победителя

**Этап 2:**
  
Игра до 3 побед


Ввод для "Этап 1":

```
Камень
```

Вывод:

```
Компьютер: Ножницы
Игрок: Камень
Победа: Игрок
```

````python
from random import randint


choices = ['камень', 'ножницы', 'бумага']
results = ['Победил игрок', 'Победил компьютер', 'Ничья']
score = 0
ai_score = 0
while max(score, ai_score) < 3:
  print('Счёт: %d:%d' % (score, ai_score))
  choice = input().lower()
  if choice not in choices:
    choice = input('Неверная команда, повторите ввод: ').lower()
  choice = choices.index(choice)
  ai_choice = randint(0, 2)
  if (choice + 1) % 3 == ai_choice:
    result = results[0]
    score += 1
  elif choice == ai_choice:
    result = results[2]
  else:
    result = results[1]
    ai_score += 1
  print('Компьютер: %s\nИгрок: %s\n%s\n' % (choices[ai_choice], choices[choice], result))
print('Игра окончена,', results[ai_score>score])
````

**Этап 1**
Создать саму игру (игра должна иметь привлекательный интерфейс, соотвесвующий классической висилице)

**Этап 2:**
Ввести подсчет количества очков и ограничения на попытки

**Этап 3:** Добавить возможность сразу угадать слово

````python
from random import randint


words = ['ромб', 'королева', 'гуманизм', 'деменция', 'государство', 'карась', 'стул', 'гидроэлектростанция', 'молекула', 'радость', 'наука', 'монитор', 'лампа', 'цвет', 'облако', 'кувшин', 'алфавит', 'символ', 'радио', 'аномалия', 'восприятие', 'сообщение', 'генотип', 'соль', 'эволюция', 'митохондрия', 'мембрана']
degradation_stadies = ['\n\n\n\n\n\n\n\n', '\n\n\n\n\n\n\n|', '\n\n\n\n\n\n\n|       |', '\n\n\n\n\n\n=========\n|       |', '\n       | \n       | \n       | \n       | \n       | \n=========\n|       |', '  +----+ \n       | \n       | \n       | \n       | \n       | \n=========\n|       |', '  +----+ \n  |    | \n  O    | \n       | \n       | \n       | \n=========\n|       |', '  +----+ \n  |    | \n  O    | \n  |    | \n       | \n       | \n=========\n|       |', '  +----+ \n  |    | \n  O    | \n /|    | \n       | \n       | \n=========\n|       |', '  +----+ \n  |    | \n  O    | \n /|\   | \n       | \n       | \n=========\n|       |', '  +----+ \n  |    | \n  O    | \n /|\   | \n /     | \n       | \n=========\n|       |', '  +----+ \n  |    | \n  O    | \n /|\   | \n / \   | \n       | \n=========\n|       |']
print('Вы играете в игру Виселица!\n')
mistakes = 0
word = words[randint(0, len(words)-1)]
pred = ['_'] * len(word)
while mistakes < 11:
  if ''.join(pred) == word:
      print('Вы угадали слово!')
      exit(0)
  print(degradation_stadies[mistakes])
  print(*pred, sep='')
  literal = input('Введите букву или слово целиком: ')
  if len(literal) > 1:
    if literal == word:
      print('Вы угадали слово!')
      break
    else:
      mistakes = 11
      break
  if literal in word:
    for j in [i for i, x in enumerate(word) if x == literal]:
      pred[j] = literal
  else:
    print('Такой буквы нет\n')
    mistakes += 1
if mistakes >= 11:
 print(degradation_stadies[11], '\nИгра окончена')
else:
  print('      \n |  | \n |  | \n_    _\n ____ ')
````
