# Задание 1

Задание:

Создайте класс с именем Rectangle который имеет:
- Атрибуты ширины и высоты.
- Метод расчета площади.
- Метод расчета периметра.
- Метод отображения размеров прямоугольника.

Создайте экземпляр класса Rectangleи продемонстрируйте его функциональность.

```python
class Rectangle:

  def __init__(self, width_value, height_value):
    self.width = width_value
    self.height = height_value

  def get_square(self):
    return self.width * self.height

  def get_perimeter(self):
    return 2*self.width + 2*self.height

  def get_dimensions(self):
    return (self.width, self.height)


rectangle = Rectangle(width_value=4, height_value=2)
print('Площадь прямоугольника rectangle:', rectangle.get_square())
print('Периметр прямоугольника rectangle:', rectangle.get_perimeter())
print('Размеры прямоугольника rectangle:', *rectangle.get_dimensions())
```

# Задание 2

Задание: Создайте мини версию банковской системы:


Инструкции:

1. Создайте класс BankAccountсо следующими атрибутами:
    - account_holder -  владелец счета
    - balance - баланс счета

2. Реализуйте следующие методы:
    - Метод для инициализации владельца счета: имя владельца счета и установите начальный баланс на 0.
    - deposit(amount): Добавьте указанную сумму к балансу.
    - withdraw(amount): Вычесть указанную сумму из баланса, если средств достаточно; в противном случае вывести предупреждение.
    - get_balance(): Возврат текущего баланса.


Создайте объект класса и продемонстрируйте его возможности

```python
class BankAccount:

    def __init__(self, account_holder):
        self.account_holder = account_holder
        self.balance = 0.0

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Сумма {amount} успешно добавлена на счет." +
                  f" Текущий баланс: {self.get_balance()}")
        else:
            print("Сумма для пополнения должна быть положительной.")

    def withdraw(self, amount):
        if amount > 0:
            if amount <= self.balance:
                self.balance -= amount
                print(f"Сумма {amount} успешно снята с счета." +
                      f" Текущий баланс: {self.get_balance()}")
            else:
                print(f"Недостаточно средств для снятия." +
                      f" Текущий баланс: {self.get_balance()}")
        else:
            print("Сумма для снятия должна быть положительной.")

    def get_balance(self):
        return self.balance


account = BankAccount("Иван Иванов")


account.deposit(1000)
account.withdraw(300)
account.withdraw(800)
account.deposit(-50)
account.withdraw(-20)
print(f"Итоговый баланс: {account.get_balance()}")
```

# Задание 3

Возьмите код и задание (Рыцарь и дракон) из предыдущей практики и реализуйте его с применением классов

```python
from random import randint


class Character:

    def __init__(self, name, armor, weapon, damage, health, heals):
        self.name = name
        self.armor = armor
        self.weapon = weapon
        self.damage = damage
        self.health = health
        self.heals = heals

    def attack(self, target, heavy=False):
        damage_multiplier = 1.5 if heavy else 1
        temp_damage = ((self.damage + randint(-self.damage, self.damage) / 10)
        * damage_multiplier)
        target.health -= temp_damage
        return temp_damage

    def heal(self):
        if self.heals > 0:
            self.health += 150
            self.heals -= 1
            return True
        return False


class Dragon:

    def __init__(self, name, health, damage):
        self.name = name
        self.health = health
        self.damage = damage

    def attack(self, target):
        temp_damage = ((self.damage + randint(-self.damage, self.damage) / 10)
        * (1 - target.armor / 100))
        target.health -= temp_damage
        return temp_damage


def create_character():
    ch_name = input('Введите имя рыцаря: ')
    ch_armor = int(input('Введите количество брони рыцаря числом: '))
    ch_weapon = input('Введите оружие рыцаря: ')
    ch_damage = int(input('Введите его урон: '))
    ch_hp = int(input('Введите здоровье рыцаря: '))
    ch_heal = int(input('Введите количество хилок: '))
    character = Character(ch_name, ch_armor, ch_weapon,
                          ch_damage, ch_hp, ch_heal)
    print('Рыцарь создан\n')
    return character


def create_dragon():
    dragon_name = input('Введите имя дракона: ')
    dragon_hp = int(input('Введите здоровье дракона: '))
    dragon_damage = int(input('Введите урон дракона: '))
    dragon = Dragon(dragon_name, dragon_hp, dragon_damage)
    print('Дракон создан\n')
    return dragon


def battle(character, dragon):
    while dragon.health > 0 and character.health > 0:
        print('Здоровье дракона:', dragon.health)
        print('Здоровье рыцаря:', character.health, '\n')
        print('Вы можете провести легкую атаку или тяжелую атаку.\n')
        print('1 - легкая атака')
        print('2 - тяжелая атака')

        choice = input()
        if choice == '1':
            damage_done = character.attack(dragon)
            print('Проведена легкая атака, нанесено', damage_done, 'урона.')
        elif choice == '2':
            damage_done = character.attack(dragon, heavy=True)
            print('Проведена тяжелая атака, нанесено', damage_done, 'урона.')
        else:
            print('Неверный ввод\n')
            continue

        if dragon.health > 0:
            dragon_damage_taken = dragon.attack(character)
            print('Вы получили', dragon_damage_taken, 'урона.\n')

        if character.heals:
            print('У вас есть', character.heals, 'хилок, вы можете '+
            'восстановить 150 здоровья, введя "heal".')
            choice = input()
            if choice.lower() == 'heal':
                character.heal()
                print('Восстановлено 150 здоровья.\n')
            else:
                print("Вы не использовали хилку.\n")
        else:
            print('Вы не можете восстановить здоровье.\n')

    if character.health <= 0:
        print('YOU DIED')
    else:
        print('Поздравляем, вам удалось одолеть дракона', dragon.name)


def main():
    print('Создайте рыцаря\n')
    character = create_character()
    print('Создайте дракона\n')
    dragon = create_dragon()
    print('Вы - рыцарь', character.name, ', который был отправлен королем',
          'королевства Ларион победить дракона',
          dragon.name, 'терроризирующего королевство на протяжении многих лет.')
    print('Чтобы вступить в бой с ним, введите "В бой!"')
    choice = input()
    print()
    if choice.lower() == 'в бой!':
        print('Битва начинается\n')
        battle(character, dragon)
    else:
        print('Вы позорно бежите')


main()
```
