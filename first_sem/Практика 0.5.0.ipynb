# **Задание 4**




Дан словарь, необходимо написать функцию меняющую ключ и значение местами

Дано:


```
{"a": 1, "b": 2, "c": 3}
```

Вывод:

```
{1: 'a', 2: 'b', 3: 'c'}
```



````python
def uno_reverse(d):
  k = list(d.keys())
  v = list(d.values())
  d1 = {}
  for i in range(len(k)):
    d1[v[i]] = k[i]
  return d1


inpt = {"a": 1, "b": 2, "c": 3}
print(uno_reverse(inpt))
````

# **Задание 5**

Дан список слов, неограниченной длинны, сформируйте словарь, где в качестве ключа будет слово, а в качестве значения количество символов

**Критерии**


*   Словарь необходимо отсортировать по убыванию количества элементов в списке.
*   Подсчет элементов должен быть реализован в отдельной функции
*   Сортировка пары `ключ:значение` должна быть реализована также в виде отдельной функции




Дано:
```
['apple','banana','orange','apple','apple','banana']
```


Вывод:
```
{'apple':3, 'banana': 2, 'orange': 1}
```

***Запрещено:***

*   Использование готовых функций для сортировки словаря
*   Использование готовых функций для подсчета элементов



````python
def counter(inpt):
  temp = set(inpt)
  out = {}
  for i in temp:
    c = 0
    for j in inpt:
      if i == j:
        c += 1
    out[i] = c
  return out


def get_key(d, value):
    for k, v in d.items():
        if v == value:
            return k


def dict_sorter(inpt):
  out = {}
  while inpt.values():
    out[get_key(inpt, max(list(inpt.values())))] = max(list(inpt.values()))
    inpt.pop(get_key(inpt, max(list(inpt.values()))))
  return out


words = ['apple','banana','orange','apple','apple','banana']
print(dict_sorter(counter(words)))
````

````python
def aged(inpt):
  for person in inpt.keys():
    if inpt[person]["age"] > 30:
      print(person, end=' ')
  print('\n')
  return


def cities(inpt):
  cities_list = list()
  for i in inpt.values():
    if i["city"] not in cities_list:
      cities_list.append(i["city"])
  for i in cities_list:
    c = 0
    for j in inpt.values():
      if j["city"] == i:
        c += 1
    print(i, ': ', c, ' чел.', sep='')
  print()
  return


def get_key(d, value):
    for k, v in d.items():
        if v == value:
            return k


def occupations(inpt):
  occ_list = list()
  for i in inpt.values():
    if i["occupation"] not in occ_list:
      occ_list.append(i["occupation"])
  for i in occ_list:
    temp = []
    for j in inpt.values():
      if j["occupation"] == i:
        temp.append(get_key(inpt, j))
    print(i, ':', *temp, sep=' ')
  print()
  return


aged(people_info)
cities(people_info)
occupations(people_info)
````

# **Задание 7**

Задание: Разработка системы отзывов о предметах

Описание: Создать программу на Python для хранения и управления отзывами о предметах учебного курса. Программа должна позволять пользователям добавлять, просматривать и удалять отзывы, а также вычислять средний балл по заданному предмету.

**Функционал:**

*   Добавление отзыва и оценки:
   *   Пользователь может ввести название предмета, оценку (от 1 до 5) и текст отзыва.
   *   Отзывы должны храниться в структуре данных (например, словаре), где ключом будет название предмета, а значением - список отзывов (каждый отзыв может хранить оценку и комментарий).
*   Просмотр отзывов и оценок:
   *   Пользователь может запросить отзывы для указанного предмета.
   *   Если для указанного предмета есть отзывы, программа должна отобразить список всех отзывов и соответствующих оценок.
*   Удаление отзыва:
   *   Пользователь может удалить отзыв по индексу. Необходимо заранее уведомить пользователя о том, какие отзывы доступны для удаления.
   *   Программа должна обработать ситуацию, когда индекс введен неправильно.
*   Вычисление среднего балла по предмету:
   *   Пользователь может ввести название предмета, и программа должна вычислить и вывести средний балл по всем отзывам для этого предмета.
   *   Если отзывов нет, программа должна сообщить об этом.


**Критерии:**

*   Код должен быть оформлен в виде функций
*   Необходимо обрабатывать неправильный ввод пользователя
*   Должны быть комментарии к функциям
*   Присутсвует весь дополнительный функционал



**Опционально:**

Предлагаю вам добавить свои критерии оценки или вопросы, на которые должен ответить студент, чтобы оценить пару




````python
def add_review():
  subject = input('Введите название предмета: ')
  try:
    grade = int(input('Введите оценку от 1 до 5: '))
  except:
    print('Некорректное значение, возврат в меню\n')
    return
  if grade < 1 or grade > 5:
    print('Некорректное значение, возврат в меню\n')
    return
  review = input('Введите текст отзыва: ')
  global reviews
  if subject not in list(reviews.keys()):
    reviews[subject] = []
  reviews[subject].append((grade, review))
  print('Отзыв добавлен\n')
  return


def remove_review():
  subject = input('Введите название предмета: ')
  print()
  global reviews
  if subject not in list(reviews.keys()):
    print('Отсутствуют отзывы по данному предмету\n')
    return
  reviews_temp = reviews[subject]
  c = 1
  for i in reviews_temp:
    print('Индекс: ', c, '\n', 'Отзыв: ', i[1], '\n', sep='')
    c += 1
  try:
    ind = int(input('Введите индекс отзыва для удаления: '))
  except:
    print('Некорректное значение, возврат в меню\n')
    return
  if ind > len(reviews_temp):
    print('Некорректное значение, возврат в меню\n')
    return
  reviews[subject].pop(ind-1)
  print('Отзыв удален\n')
  return


def check_reviews():
  subject = input('Введите название предмета: ')
  print()
  global reviews
  if subject not in list(reviews.keys()):
    print('Отсутствуют отзывы по данному предмету\n')
    return
  reviews_temp = reviews[subject]
  for i in reviews_temp:
    print('Оценка: ', i[0], '\n', 'Отзыв: ', i[1], '\n', sep='')
  return


def get_avg():
  subject = input('Введите название предмета: ')
  print()
  global reviews
  if subject not in list(reviews.keys()):
    print('Отсутствуют отзывы по данному предмету\n')
    return
  reviews_temp = reviews[subject]
  s = 0
  c = 0
  for i in reviews_temp:
    c += 1
    s += int(i[0])
  print('Средний балл:', s/c, '\n')
  return


def main():
  print('Список команд:\n1 - добавить отзыв\n2 - удалить отзыв\n3 - просмотреть отзывы\n4 - просмотреть средний балл\n0 - выход')
  choice = 5
  while choice != '0':
    choice = input()
    print()
    match choice:
      case '1':
        add_review()
      case '2':
        remove_review()
      case '3':
        check_reviews()
      case '4':
        get_avg()
      case '0':
        break
      case _:
        print('Некорректный ввод\n')


reviews = {}
main()
````
