# Задание 0

Создайте функцию по нахождению уникальных элементов из двух списков



```
a = [1, 2, 3, 4]
b = [3, 4, 5, 6]
```



````python
def unique(list_a, list_b):
  result = []
  for i in list_a:
    if i not in list_b:
      result.append(i)
  for i in list_b:
    if i not in list_a:
      result.append(i)
  return result


a = [1, 2, 3, 4]
b = [3, 4, 5, 6]
print(unique(a, b))
````

# Задание 1

Симулятор виртуального питомца

Цель: создать виртуальный симулятор домашних животных, в котором пользователи смогут заводить питомцев и ухаживать за ними.

Требования:

- Функция для усыновления питомца (имя, тип, возраст).
- Функция для того, чтобы покормить питомца, поиграть с ним или уложить его спать.
- Функция для отображения состояния питомца (голод, радость, энергия).

````python
def create_pet():
  pet_name = input('Введите имя нового питомца: ')
  pet_type = input('Введите тип нового питомца: ')
  pet_age = input('Введите возраст нового питомца: ')
  global pet
  pet = {'name': pet_name, 'type': pet_type, 'age': pet_age, 'hunger': 100,
         'happiness': 100, 'stamina': 100}
  print('Питомец создан\n')
  return


def feed():
  global pet
  pet['hunger'] = (pet['hunger'] + 50) % 101
  print('Питомец накормлен\n')
  return


def play():
  global pet
  pet['happiness'] = (pet['happiness'] + 25) % 101
  print('Вы поиграли с питомцем\n')
  return


def sleeping():
  global pet
  pet['stamina'] = 100
  print('Вы уложили спать питомца, его энергия восстановлена\n')
  return


def status():
  global pet
  print('Голод: ', pet['hunger'], '/100', sep='')
  print('Радость: ', pet['happiness'], '/100', sep='')
  print('Энергия: ', pet['stamina'], '/100\n', sep='')
  return


def main():
  global pet
  choice = '1'
  print('Доступные команды:')
  print('1 - создать нового питомца')
  print('2 - покормить питомца')
  print('3 - поиграть с питомцем')
  print('4 - уложить спать питомца')
  print('5 - отобразить состояние питомца')
  print('0 - выход\n')
  while choice != '0':
    choice = input()
    match choice:
      case '0':
        break
      case '1':
        print()
        create_pet()
      case '2':
        print()
        feed()
      case '3':
        print()
        play()
      case '4':
        print()
        sleeping()
      case '5':
        print()
        status()
      case _:
        print('Неверный ввод\n')
    pet['hunger'] -= 10
    pet['happiness'] -= 10
    pet['stamina'] -= 5


pet = {}
main()
````

# Задание 2

Рыцарь и дракон

Цель: создать небольшую игру, в которой вам необходимо играть за рыцаря и сразиться с драконом

Требования:

- Создание персонажа (имя, информация о доспехах, оружии, урон, здоровье)
- Управление персонажем и мини сюжет
- Создание дракона (Имя, информация о здоровье и уроне)
- Боевая система (нанесение и получение урона, урон должен быть случайным в заданном диапазоне)
- Реализовать бой между драконом и рыцарем

````python
from random import randint


def create_character():
  global character
  ch_name = input('Введите имя рыцаря: ')
  ch_armor = int(input('Введите количество брони рыцаря числом: '))
  ch_weapon = input('Введите оружие рыцаря: ')
  ch_damage = int(input('Введите его урон: '))
  ch_hp = int(input('Введите здоровье рыцаря: '))
  ch_heal = int(input('Введите количество хилок: '))
  character = {'name': ch_name, 'armor': ch_armor, 'weapon': ch_weapon,
               'damage': ch_damage, 'health': ch_hp, 'heals': ch_heal}
  print('Рыцарь создан\n')
  return


def create_dragon():
  global dragon
  dragon_name = input('Введите имя дракона: ')
  dragon_hp = int(input('Введите здоровье дракона: '))
  dragon_damage = int(input('Введите урон дракона: '))
  dragon = {'name': dragon_name, 'health': dragon_hp, 'damage': dragon_damage}
  print('Дракон создан\n')
  return


def battle():
  global dragon, character
  while dragon['health']>0 and character['health']>0:
    print('Здоровье дракона:', dragon['health'])
    print('Здоровье рыцаря:', character['health'], '\n')
    print('Вы можете провести легкую атаку с последующей ', sep='', end='')
    print('попыткой уворота от атаки дракона ', sep='', end='')
    print('или тяжелую атаку, но без возможности ', sep='', end='')
    print('увернуться.\n')
    print('1 - легкая атака')
    print('2 - тяжелая атака')
    choice = input()
    match choice:
      case '1':
        temp_damage = character['damage'] + randint(-1*character['damage'],
                                                    character['damage'])/10
        dragon['health'] = dragon['health'] - temp_damage
        print('Проведена легкая атака, нанесено',temp_damage,'урона.')
        avoiding = randint(0, 100)/100
        if avoiding >= 0.7:
          print('Вы увернулись от атаки дракона.\n')
        else:
          temp_damage = (dragon['damage'] + randint(-1*dragon['damage'],
                         dragon['damage'])/10) * (1 - character['armor']/100)
          character['health'] = character['health'] - temp_damage
          print('У вас не получилось увернуться, и вы получили', end=' ')
          print(temp_damage,'урона.\n')
      case '2':
        temp_damage = (character['damage'] + randint(-1*character['damage'],
                       character['damage'])/10) * 1.5
        dragon['health'] = dragon['health'] - temp_damage
        print('Проведена тяжелая атака, нанесено',temp_damage,'урона.')
        temp_damage = (dragon['damage'] + randint(-1*dragon['damage'],
                       dragon['damage'])/10) * (1 - character['armor']/100)
        character['health'] = character['health'] - temp_damage
        print('Вы получили', temp_damage, 'урона.\n')
      case _:
        print('Неверный ввод\n')
    if character['heals']:
      print('У вас есть', character['heals'],'хилок, вы можете ', end='')
      print('восстановить 150 здоровья, введя \"heal\".')
      choice = input()
      if choice == 'heal':
        character['heals'] = character['heals'] - 1
        character['health'] = character['health'] + 150
        print('Восстановлено 150 здоровья.\n')
      else:
        print()
        continue
    else:
      print('Вы не можете восстановить здоровье.\n')
  if character['health']<=0:
    print('YOU DIED')
    return
  else:
    print('Поздравляем, вам удалось одолеть дракона', dragon['name'])
    return


def main():
  global dragon, character
  print('Создайте рыцаря\n')
  create_character()
  print('Создайте дракона\n')
  create_dragon()
  print('Вы - рыцарь ', character['name'], ', который был ', sep='', end='')
  print('отправлен королем королевства Ларион победить ', sep='', end='')
  print('дракона ', dragon['name'], ', терроризирующего ', sep='', end='')
  print('королевство на протяжении многих лет.', sep='')
  print('Чтобы вступить в бой с ним, введите \"В бой!\"')
  choice = input()
  print()
  if choice.lower() == 'в бой!':
    print('Битва начинается\n')
    battle()
  else:
    print('Вы позорно бежите')
    return


dragon = {}
character = {}
main()
````

\# Задание 3


Цель - создать менеджера команды Pokémon, который позволит пользователям:

- Добавлять покемонов в свою команду. (если такого покемона еще нет в команде)
- Удалять покемонов из их команды.
- Просматривать подробную информацию обо всех покемонах в команде.
- Находить покемона по имени.
- Устраивать тренировочный бой между двумя покемонами

Для данной задачи используйте: https://pokeapi.co/


````python
import requests


def get_pokemons_list():
  url = 'https://pokeapi.co/api/v2/pokemon'
  responce = requests.get(url)
  data = responce.json()
  data = [x for x in data['results']]
  names = []
  for i in data:
    names.append(i['name'])
  return names


def get_info(name):
  url = 'https://pokeapi.co/api/v2/pokemon'
  responce = requests.get(url + '/' + name)
  data = responce.json()
  data_abilities = data['abilities']
  data_types = data['types']
  abilities = []
  types = []
  for i in data_abilities:
    abilities.append(i['ability']['name'])
  for i in data_types:
    types.append(i['type']['name'])
  result = {'name': name, 'types': types, 'weight': data['weight'],
            'height': data['height'], 'abilities': abilities}
  return result


def add_pokemon():
  global team, pokemons_list
  print()
  print(*pokemons_list, sep=', ')
  poke_name = input('Введите имя покемона из списка: ')
  if poke_name not in pokemons_list:
    print('Неверное имя\n')
  else:
    temp_info = get_info(poke_name)
    team[poke_name] = temp_info
    print(poke_name, 'добавлен в команду\n')
  return


def remove_pokemon():
  global team
  names = list(team.keys())
  print('Состав команды: ', end='')
  print(*names, sep=', ')
  name_to_remove = input('Введите имя покемона, которого хотите убрать: ')
  if name_to_remove not in names:
    print('Неверное имя\n')
  else:
    del team[name_to_remove]
    print(name_to_remove, 'убран из команды\n')
  return


def get_pokemon_info():
  global team
  names = list(team.keys())
  print('Состав команды: ', end='')
  print(*names, sep=', ')
  name_to_get_info = input('Введите имя покемона, о котором хотите узнать: ')
  if name_to_get_info not in names:
    print('Неверное имя\n')
  else:
    print(team[name_to_get_info])
    print()
  return


def train_battle():
  global team
  names = list(team.keys())
  print('Состав команды: ', end='')
  print(*names, sep=', ')
  print('Введите через запятую два покемона для боя')
  pokemon1, pokemon2 = map(str, input().split(', '))
  if pokemon1 not in names or pokemon2 not in names:
    print('Неверное имя\n')
  else:
    print('\nНачинается тренировочный бой')
    print('Способности ', pokemon1, ':', sep='', end=' ')
    print(*team[pokemon1]['abilities'], sep=', ')
    print('Способности', pokemon2, ':', sep='', end=' ')
    print(*team[pokemon2]['abilities'], sep=', ')
    print('Тренировочный бой завершен\n')
  return


def main():
  global pokemons_list
  pokemons_list = get_pokemons_list()
  print('Вы можете:')
  print('1 - добавить покемона в команду')
  print('2 - убрать покемона из команды')
  print('3 - просмотреть информацию о покемонах в команде')
  print('4 - начать тренировочный бой')
  print('0 - выйти\n')
  choice = ''
  while choice != '0':
    choice = input()
    match choice:
      case '1':
        add_pokemon()
      case '2':
        remove_pokemon()
      case '3':
        get_pokemon_info()
      case '4':
        train_battle()
      case '0':
        break
      case _:
        print('Неверный ввод\n')


pokemons_list = []
team = {}
main()
````
